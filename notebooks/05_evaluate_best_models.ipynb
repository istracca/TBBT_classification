### Import utility functions and libraries

```python
import os
import sys
import itertools

# Adding 'src' directory to the system path
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(os.path.join(project_root, 'src'))

from utils import print_model_results, evaluate_model_on_test_set
```

### Classifier only (Frozen SBERT)

```python
tasks = ["all", "Sheldon_Penny", "Sheldon_Leonard"]
losses = ["cross_entropy", "weighted_cross_entropy"]

for task, loss in itertools.product(tasks, losses):

    log_dir = f"../metrics/classifier_only/{task}/{loss}"
    print(f"Results for {task} with {loss}:")

    # Getting the best parameters from the model results
    best_params = print_model_results(log_dir=log_dir, patience=10, max_epochs=1000, model_type="classifier_only")
    print("\n" + "="*50 + "\n")
    
    if best_params:
        # Extracting parameters from the best_params dictionary
        lr = float(best_params["learning_rate"])
        wd = float(best_params["weight_decay"])

        # Formatting learning rate and weight decay for display
        def format_param(val):
            return f"{val:.0e}" if val < 1e-4 else f"{val:.4f}".rstrip('0').rstrip('.')
        lr_str = format_param(lr)
        wd_str = format_param(wd)

        # Formatting dropout rate for display
        dropout_val = best_params["dropout_rate"]
        dropout_str = 0 if dropout_val == 0.0 else dropout_val

        # Evaluating the model on the test set with the best parameters
        evaluate_model_on_test_set(
            task=task,
            loss=loss,
            lr=lr_str,
            dropout=dropout_str,
            wd=wd_str,
            model_type="classifier_only"
        )
```

### LoRA

```python
tasks = ["all", "Sheldon_Penny", "Sheldon_Leonard"]
losses = ["cross_entropy", "weighted_cross_entropy"]

for task, loss in itertools.product(tasks, losses):

    log_dir = f"../metrics/LoRA/{task}/{loss}"
    print(f"Results for {task} with {loss}:")

    # Getting the best parameters from the model results
    best_params = print_model_results(log_dir=log_dir, patience=3, max_epochs=20, model_type="lora")
    print("\n" + "="*50 + "\n")

    if best_params:
        # Extracting parameters from the best_params dictionary
        lr = float(best_params["learning_rate"])
        rank = int(best_params["rank"])
        alpha = int(best_params["alpha"])
        wd = float(best_params["weight_decay"])

        # Formatting learning rate and weight decay for display
        def format_param(val):
            if val == 0:
                return 0
            return f"{val:.0e}" if val < 1e-4 else f"{val:.4f}".rstrip('0').rstrip('.')
        lr_str = format_param(lr)
        wd_str = format_param(wd)

        # Formatting dropout rate for display
        dropout_val = best_params["dropout_rate"]
        dropout_str = 0 if dropout_val == 0.0 else dropout_val

        # Evaluating the model on the test set with the best parameters
        evaluate_model_on_test_set(
            task=task,
            loss=loss,
            rank=rank,
            alpha=alpha,
            lr=lr_str,
            dropout=dropout_str,
            wd=wd_str,
            model_type="lora"
        )
```
